## Merged into a Reccuring Function

```python
import pandas as pd
import datetime
import time
```

```python
# Run once to get original table
df = pd.read_html('https://www.skysports.com/premier-league-table')
table = df[0].drop(columns=['Last 6'])
table['Date'] = datetime.date.today()
table.to_csv('table.csv', index=False)
```

```python
def addnewstandings() :
    df = pd.read_html('https://www.skysports.com/premier-league-table')
    table = df[0].drop(columns=['Last 6'])
    table['Date'] = datetime.date.today()
    nohead = pd.read_csv('table.csv')
    append = pd.DataFrame.append(nohead, table)
    append.to_csv('table.csv', index=False)
```

```python
while(True):
    addnewstandings()
    time.sleep(60*60*24*7) #Updates every week
```
